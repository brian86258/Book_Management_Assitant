```python
import requests
import matplotlib.pyplot as plt
import json
```

```python
from PIL import Image
from io import BytesIO
```

```python
isbn = 9789861343242
# 9789861850719
url = 'http://static.findbook.tw/image/book/{}/medium'.format(isbn)
r = requests.get(url = url)
im = Image.open(BytesIO(r.content))

plt.imshow(im)
plt.show()
```

# GOOGLE BOOK API

```python

# isbn = 9781292018195
# isbn = 1292018194
# isbn = 9789863984412
isbn = 9789866281860
```

```python
search_url = 'https://www.googleapis.com/books/v1/volumes?q=isbn:{}'.format(isbn)
r = requests.get(url = search_url)

print(r)
# print(r.text)

```

```python
result = json.loads(r.text)
```

```python
# result['items'][0]['volumeInfo']
```

```python
categories = result['items'][0]['volumeInfo']['categories']
categories
```
